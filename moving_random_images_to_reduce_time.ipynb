```python
from PIL import Image
import scipy
from scipy import misc
import numpy as np
import os 
import random
import shutil
import os.path


inpath = './flairdata/preprocess_Data/'

outpath = './reduced_flairdata/'
files = os.listdir(inpath)

num_to_select = len(files)
list_of_random_files = random.sample(files, num_to_select)


        #if thing in some_list: some_list.remove(thing)

total_i = 5000
i=0

while i<total_i:
    ori_name = list_of_random_files[i]
    print(ori_name)
    if 'seg' in ori_name: 
        pair_file = ori_name.replace('seg', 'flair')
    elif 'flair' in ori_name:
        pair_file = ori_name.replace('flair', 'seg')
    else:
        print('Something goes wrong')
        break
    print('this is ori: ' + ori_name)
    print('this is pair_file: ' + pair_file)
    if os.path.exists(inpath+ori_name):
        if os.path.exists(inpath+pair_file):
            os.rename(inpath+ori_name, outpath+ori_name)
            os.rename(inpath+pair_file, outpath+pair_file)
            list_of_random_files.remove(ori_name)
            list_of_random_files.remove(pair_file)
    i+=1
    







```
