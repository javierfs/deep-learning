```python
from PIL import Image
import scipy
from scipy import misc
import numpy as np
import os
path = './Data/test/'
outpath = './img/test/'
files = os.listdir(path)

for file in files:
    print(file)
    data = np.load(path + file)
    data = (data - data.min()) / (data.max() - data.min()).astype(np.float32)
    save_filename = file.replace('.nii.npy', '')
    scipy.misc.imsave(outpath+save_filename+'.jpg', data)
```
